```python
import numpy as np
```

```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import difflib
```

```python
movie_dataset=pd.read_csv('movies.csv')
```

```python
movie_dataset.head()
```

```python
movie_dataset.columns
```

```python
selected_feature=['genres','keywords','tagline','title','cast', 'director',]
```

```python
print(selected_feature)
```

```python
for feature in selected_feature:
    movie_dataset[feature]=movie_dataset[feature].fillna(' ')
```

```python
combine_feature=movie_dataset['genres']+' '+movie_dataset['keywords']+' '+movie_dataset['tagline']+' '+movie_dataset['title']+' '+movie_dataset['cast']+' '+movie_dataset['director']
```

```python
combine_feature.isnull().sum()
```

```python
vector=TfidfVectorizer()
```

```python
vector_feature=vector.fit_transform(combine_feature)
```

```python
print(vector_feature)
```

```python
similar=cosine_similarity(vector_feature)
```

```python
similar
```

```python
similar.shape
```

```python
user_movie='iron man'
```

```python
list_of_the_title=movie_dataset['title'].tolist()
```

```python
# print(list_of_the_title)
```

```python
find_close_match=difflib.get_close_matches(user_movie,list_of_the_title)
```

```python
print(find_close_match)
```

```python
close_match=find_close_match[0]
```

```python
print(close_match)
```

```python
index_of_the_movie=movie_dataset[movie_dataset.title==close_match]['index'].values[0]
```

```python
index_of_the_movie
```

```python
movie_dataset.iloc[[68]]
```

```python
similarity_score=list(enumerate(similar[index_of_the_movie]))
```

```python
len(similarity_score)
```

```python
stored_similar_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)
```

```python
print('Movie suggestion for you:\n')
```

```python
print('Movie suggestion for you:\n')
i=1
for movie in stored_similar_movie:
    index=movie[0]
    title_from_index = movie_dataset[movie_dataset.index == index]['title'].values[0]

    if (i<30):
        print(i,'.',title_from_index)
        i+=1
```

```python
print('Movie suggestion for you:\n')

i = 1
for movie in stored_similar_movie:
    index = movie[0]
    title_from_index = movie_dataset.loc[index, 'title']

    if i < 30:
        print(i, '.', title_from_index)
        i += 1
```
